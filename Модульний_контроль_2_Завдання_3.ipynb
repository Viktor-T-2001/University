```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_excel('МКР2 Варіант 4 авіаперевезення.xlsx')
df = df.dropna(subset=['Verified'])
columns_to_keep = ['Type Of Traveller', 'Seat Type']
columns_to_drop = [col for col in df.columns if col not in columns_to_keep]
df.drop(columns=columns_to_drop, inplace=True)
df.tail(2)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
# Використовуємо pd.get_dummies() для створення бінарних стовпців
dummies = pd.get_dummies(df['Type Of Traveller'])

# Об'єднуємо з початковим DataFrame
df = pd.concat([df, dummies], axis=1)
df.tail()
```

```python
# Використовуємо pd.get_dummies() для створення бінарних стовпців
dummies = pd.get_dummies(df['Seat Type'])

# Об'єднуємо з початковим DataFrame
df = pd.concat([df, dummies], axis=1)
df.tail()
```

```python
df.drop(columns=['Type Of Traveller','Seat Type'], inplace=True)
df.tail(2)
```

```python
# Building the model
frq_items = apriori(df, min_support = 0.05, use_colnames = True)

# Collecting the inferred rules in a dataframe
rules = association_rules(frq_items, metric ="confidence", min_threshold = 0.1)
rules = rules.sort_values(['confidence', 'support'], ascending =[False, False])
rules
```

### **Побудова різних видів графіків технічного аналізу за реальною статистикою  ринку цінних паперів**

Отримуємо дані зі щоденною періодичністю за останні 50 днів для акції ORCL. Датасет включає в себе ціни відкриття, закриття, максимальну та мінімальну ціну за день, обсяги торгів, а також розмір виплачених дивидендів (якщо виплата дивідендів не відбувалася, то для таких днів значення у цьому стовпці = 0), та кофецієнт для спліту акцій (якщо спліт відбувся у вказаний день, інакше 0)

```python
import yfinance
ORCL = yfinance.Ticker('ORCL')
ORCL_data = ORCL.history(period='50d')
```

```python
ORCL_data.drop(['Dividends', 'Stock Splits'], axis=1, inplace=True)
ORCL_data
```

Перевіряємо, що в кожному стовпці справді по 50 значень (count):

```python
ORCL_data.describe()
```

#### **Лінійні графіки**

Почнемо з побудови лінійних графіків з усіма цінами (закриття, відкриття, мінімальна, та максимальна) та з об'ємом торгів:

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

fig = go.Figure()
fig = make_subplots(rows=2, cols=1, shared_xaxes=True,
               vertical_spacing=0.06, subplot_titles=('ORCL', 'Об`єм торгів'),
               row_width=[0.2, 0.7])

fig.add_trace(go.Scatter(x=ORCL_data.index, y=ORCL_data['Close'],
                    mode='lines',
                    name='ORCL - ціна закриття',
                    line=dict(color='red', width=1.5)),
                    row=1, col=1
                    )
fig.add_trace(go.Scatter(x=ORCL_data.index, y=ORCL_data['Open'],
                    mode='lines',
                    name='ORCL - ціна відкриття',
                    line=dict(color='green', width=1.5)),
                    row=1, col=1
                    )
fig.add_trace(go.Scatter(x=ORCL_data.index, y=ORCL_data['High'],
                    mode='lines',
                    name='ORCL - максимальна ціна',
                    line=dict(color='royalblue', width=1.5)),
                    row=1, col=1
                    )
fig.add_trace(go.Scatter(x=ORCL_data.index, y=ORCL_data['Low'],
                    mode='lines',
                    name='ORCL - мінімальна ціна',
                    line=dict(color='purple', width=1.5)),
                    row=1, col=1
                    )
fig.add_trace(go.Bar(x=ORCL_data.index, y=ORCL_data['Volume'],
                    name='ORCL - об`єм торгів',
                    marker_color='coral'),
                    row=2, col=1
                    )

fig.show()
```

Як ми бачимо з попереднього графіку, вихідні дні, хоча для них і немає значень по цінах акції, адже біржі в ці дні не працюють, все одно не відображаються. Через це, наприклад, на графіку з об'ємами торгів є пропуски. Нам потрібно додати умову до графіку, за якою відображалися б тільки робочі дні.

Також не будемо відображати окремі святкові дні, під час яких біржі також зачинені. #### **Японські свічки**

Тепер перейдемо до побудови біржової діаграми "Японські свічки", яка відображатиме відразу всі 4 ціни - відкриття, закриття, максимальну, та мінімальну.

#### **Графік хрестики-нулики**

Інша назва - пункто-цифровий графік, відображає зміну ціни за рахунок символів O - падіння ціни, X - зростання ціни. Два основних параметри - це величина порогу (для графіку нижче це значення встановлено як 1) яка визначає мінімально необхідний рівень зміни ціни, та параметр розвороту (для графіку нижче це значення встановлено як 2), який визначає скільки порогових змін потрібно для зміни тренду.

```python
import mplfinance as mpf

mpf.plot(ORCL_data,type='pnf', pnf_params=dict(box_size=1,reversal=2))
```

#### **Графік Ренко**

Цей графік відображає зміну ціни (якщо вона вище певного наперед встановленого рівня, для графіку нижче таким пороговим значенням встановлено значення 1) як послідовність діагонально розташованих квадратів.

```python
mpf.plot(ORCL_data,type='renko',renko_params=dict(brick_size=1))
```

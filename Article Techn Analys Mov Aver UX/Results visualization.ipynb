```python
import pandas as pd
import missingno as msno
import matplotlib.pyplot as plt
import missingno as msno
import numpy as np
```

```python
columns = ['Тип','Параметр','Результат']
df = pd.read_excel('Післявоєнні результати.xlsx', header = None, skiprows = 1, names = columns)
df.head()
```

```python
max_value = max(df["Результат"])
fig, axs = plt.subplots(1, 3, figsize=(20, 5))

plot1 = plt.subplot2grid((1, 6), (0, 0), colspan=2) 
plot2 = plt.subplot2grid((1, 6), (0, 2), colspan=2) 
plot3 = plt.subplot2grid((1, 6), (0, 4), colspan=2) 
    
plot1.plot(df["Параметр"].loc[(df['Тип']=='SMA')], df["Результат"].loc[(df['Тип']=='SMA')], color='navy', linewidth=1.5, drawstyle="steps-post")
plot1.set_ylim(0, max_value*1.05)
plot1.set_title('SMA') 
plot1.set_ylabel('\nРезультат (початкове значення капіталу = 100)')

plot2.plot(df["Параметр"].loc[(df['Тип']=='EMA')], df["Результат"].loc[(df['Тип']=='EMA')], color='navy', linewidth=1.5, drawstyle="steps-post")
plot2.set_ylim(0, max_value*1.05)
plot2.set_title('EMA')
plot2.set_xlabel('\nЗначення параметра n')

plot3.plot(df["Параметр"].loc[(df['Тип']=='LWMA')], df["Результат"].loc[(df['Тип']=='LWMA')], color='navy', linewidth=1.5, drawstyle="steps-post")
plot3.set_ylim(0, max_value*1.05)
plot3.set_title('LWMA')
  
plt.tight_layout() 
plt.show() 
```

```python
from statsmodels.graphics.tsaplots import plot_acf
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_excel('Task 1 Data.xlsx', header = 0)
df.index = df["Year"]
df = df.iloc[::-1]
```

Use the Autocorrelation function from the statsmodel library passing our DataFrame object in as the data

```python
plt.rcParams["figure.figsize"] = (20,12)
```

```python
plot_acf(x=df["The ratio of gross external debt to GDP, %"], title=' - The ratio of gross external debt to GDP')
plt.show()
```

```python
plot_acf(x=df["The ratio of gross external debt to exports, %"], title=' - The ratio of gross external debt to exports')
plt.show()
```

```python
plot_acf(x=df[~df['The ratio of the public and publicly guaranteed external debt to the budget revenues, %'].isnull()]["The ratio of the public and publicly guaranteed external debt to the budget revenues, %"], title=' - The ratio of the public and publicly guaranteed external debt to the budget revenues')
plt.show()
```

```python
plot_acf(x=df["Average maturity of external obligations (years)"], title=' - Average maturity of external obligations')
plt.show()
```

```python
plot_acf(x=df["Average interest rate on external public liabilities (%)"], title=' - Average interest rate on external public liabilities')
plt.show()
```

```python
plot_acf(x=df["The ratio of payments for servicing external debt to GDP, %"], title='The ratio of payments for servicing external debt to GDP')
plt.show()
```

```python
plot_acf(x=df[~df['The ratio of payments for servicing external debt to exports, %'].isnull()]["The ratio of payments for servicing external debt to exports, %"], title=' - The ratio of payments for servicing external debt to exports')
plt.show()
```
